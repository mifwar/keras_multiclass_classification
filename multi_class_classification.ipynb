```python
#create function to generate class label
def f(row):
    if row['variety'] == 'Setosa':
        val = 0
    elif row['variety'] == 'Versicolor':
        val = 1
    else:
        val = 2
    return val
```

```python
#turn index into label name
def reverse(idx):
    if idx == 0:
      return 'Setosa'
    elif idx == 1:
      return 'Versicolor'
    else:
      return 'Virginica'
```

```python
#load csv and clear some columns. Make sure that you have iris.csv in /content/drive/MyDrive/test_multi_class/
df = pd.read_csv('/content/drive/MyDrive/test_multi_class/iris.csv')
df['class'] = df.apply(f, axis=1)
properties = list(df.columns.values)
properties.remove('variety')
properties.remove('class')
print(properties)
```

```python
#assign dataFrame
dataX = df[properties]
dataY = df['class']

print(dataX)
print(dataY)
```

```python
#split data
trainX, testX, trainY, testY = train_test_split(dataX, dataY, test_size=0.2, random_state=0)
print(trainX.head())
print(testX.head())
print(trainY.head())
print(testY.head())
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(4, )),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(3)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(trainX, trainY, epochs=1500, batch_size=16)
test_loss, test_acc = model.evaluate(testX, testY)
print('Test accuracy:', test_acc)
print('Test loss:', test_loss)
```

```python
#test
probability_model = tf.keras.Sequential([model, tf.keras.layers.Softmax()])
predictions = probability_model.predict(testX[21:22])
print(testX[21:22])
print(predictions[0:])
print(reverse(np.argmax(predictions[0:])))
```
